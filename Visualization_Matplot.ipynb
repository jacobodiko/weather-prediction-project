```python
import pandas as pd 
import numpy as np 

import matplotlib.pyplot as plt

# Sample dataset
data = {
    'day_of_week': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'temperature': [22, 28, 19, 21, 25, 30, 27],
    'humidity': [65, 55, 80, 70, 60, 50, 75],
    'wind_speed': [10, 5, 12, 9, 15, 6, 8],
    'pressure': [1012, 1018, 1009, 1010, 1013, 1020, 1016],
    'rain': [1, 0, 1, 0, 0, 0, 1]
}

df = pd.DataFrame(data)

print(df)
```

```python
#line plot for temperature ,humidity and Windspeed

plt.figure(figsize=(10,6))

# Plot temperature, humidity, and wind speed as lines
plt.plot(df['day_of_week'], df['temperature'], label='Temperature (°C)', color='red', marker='o')
plt.plot(df['day_of_week'], df['humidity'], label='Humidity (%)', color='blue', marker='o')
plt.plot(df['day_of_week'], df['wind_speed'], label='Wind Speed (km/h)', color='green', marker='o')


# Adding labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Value')
plt.title('Weather Data Over the Week')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting rain data

plt.figure(figsize=(10, 6))


rain_days = df['rain'] == 1
plt.bar(df['day_of_week'][rain_days], df['rain'][rain_days], color='skyblue', label='Rainy Day')
plt.bar(df['day_of_week'][~rain_days], df['rain'][~rain_days], color='orange', label='No Rain')
```

```python
# Scatter plot for temperature vs. humidity
plt.scatter(df['temperature'][df['rain'] == 1], df['humidity'][df['rain'] == 1], color='blue', label='Rain', s=100)
plt.scatter(df['temperature'][df['rain'] == 0], df['humidity'][df['rain'] == 0], color='red', label='No Rain', s=100)

# Adding labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.title('Temperature vs Humidity (Rain vs No Rain)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(8, 6))
# Histogram for wind speed
plt.hist(df['wind_speed'], bins=5, color='purple', edgecolor='black')

plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.title('Distribution of Wind Speed')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))

# Box plot for pressure
plt.boxplot(df['pressure'], vert=False)

plt.xlabel('Pressure (hPa)')
plt.title('Box Plot of Pressure')
plt.grid(True)
plt.show()
```
